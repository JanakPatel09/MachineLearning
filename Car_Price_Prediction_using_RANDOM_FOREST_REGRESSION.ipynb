#Car Price Prediction using RANDOM FOREST REGRESSION

```python
#Import library

import pandas as pd
```

```python
#Load dataset from Local Directort

from google.colab import files
data = files.upload()
```

```python
#Load dataset & Summarize data

dataset = pd.read_csv('dataset.csv')
dataset = dataset.drop(['car_ID'], axis = 1)
print(dataset.shape)
print(dataset.head())
```

```python
#Splitting dataset into X & Y
#X contains numerical and text data. So, we remove taxt data and go with only numerical data.
xData = dataset.drop('price', axis = 'columns')
numData = xData.select_dtypes(exclude=['object']).columns
X = xData[numData]

Y = dataset['price']
```

```python
#Scaling the independent variables

from sklearn.preprocessing import scale
cols = X.columns
X = pd.DataFrame(scale(X))
X.columns = cols
X
```

```python
#Splitting dataset for training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
#Training our model using Random Forest

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train)
print('Training completed.')
```

```python
#Evaluationg our model

ypred = model.predict(x_test)

from sklearn.metrics import r2_score
r2score = r2_score(y_test, ypred)
print("R2Score:", r2score*100)
```
